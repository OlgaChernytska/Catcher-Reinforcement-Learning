### Part 2: Cross Entropy Method

```python
import numpy as np
import pandas as pd
from ple import PLE
from ple.games.catcher import Catcher
import matplotlib.pyplot as plt
import time 
```

```python
class LearningAgent:
    def __init__(self, actions, theta):
        self.actions = actions
        self.theta = theta
        
    def pick_action(self, obs):
        obs = list(game.getGameState().values())
        obs = np.hstack((1, obs))

        if np.dot(self.theta, obs) > 0:
            action = self.actions[0]
        else:
            action = self.actions[1]
        return action
```

```python
# initialize game
game = Catcher(width=100, height=100, init_lives=1)
p = PLE(game, fps=30, frame_skip=3, num_steps=1, force_fps=True, display_screen=False)
p.init()

# to find out how parameters perform, we need to play using them
def f(theta):
    p.reset_game()
    agent = LearningAgent(p.getActionSet(), theta)

    max_timestamps = 400
    for timestamp in range(max_timestamps):
        obs = game.getGameState()
        action = agent.pick_action(obs)
        reward = p.act(action)
        if p.game_over():
            break

    return timestamp+1
```

```python
# tracking training time
start_time = time.time()

# experiment parameters
params = {
    'theta dimension': 5, 
    'iterations': 1000,
    'batch size': 50,
    'elite fraction': 0.2
}

# initialize parameters
theta_mean = np.zeros(params['theta dimension'])
theta_std = np.ones(params['theta dimension'])

# write logs
logs = {'episode': [],
        'elite_avg_timestamp_last_20_episodes': [],
        'all_avg_timestamp_last_20_episodes': []}

dim_theta = params['theta dimension']
elite_frac = params['elite fraction']
batch_size = params['batch size']
n_iter = params['iterations']
iteration_results_elite = []
iteration_results_all = []

for iter_index in range(n_iter):
    # Sample parameter vectors
    thetas = np.random.normal(theta_mean, theta_std, (batch_size, dim_theta))
    scores = np.array([f(theta) for theta in thetas])
    # Get elite parameters
    elite_indices = np.argsort(-scores)[:int(batch_size * elite_frac)]
    elite_thetas = thetas[elite_indices, :]
    # Update theta_mean, theta_std
    theta_mean = elite_thetas.mean(axis=0)
    theta_std = elite_thetas.std(axis=0)

    iteration_results_elite.append(np.mean(scores[elite_indices]))
    iteration_results_all.append(np.mean(scores))
    

    if (iter_index + 1) % 20 == 0:
        avg_score_elite = np.mean(iteration_results_elite[len(iteration_results_elite)-20:])
        avg_score_all = np.mean(iteration_results_all[len(iteration_results_all)-20:])
        print('Episode %3d Elite mean: %3d, General mean: %3d'% (iter_index+1, avg_score_elite, avg_score_all))
    
    # write logs
    avg_score_elite = np.mean(iteration_results_elite[len(iteration_results_elite)-20:])
    avg_score_all = np.mean(iteration_results_all[len(iteration_results_all)-20:])
    logs['episode'].append(iter_index + 1)
    logs['elite_avg_timestamp_last_20_episodes'].append(avg_score_elite)
    logs['all_avg_timestamp_last_20_episodes'].append(avg_score_all)

    # convergence condition
    if avg_score_all >= 200:
        print('Early stopping. General mean over last 20 iterations is %3d' % (avg_score_all))
        break

print('Training time: %.1f minutes.' % ((time.time() - start_time) / 60))
```

```python
# save model
np.save('../model/cem_theta.npy', theta_mean)

# save logs
logs = pd.DataFrame(logs)
logs = logs[['episode','elite_avg_timestamp_last_20_episodes','all_avg_timestamp_last_20_episodes']]
logs.to_csv('../logs/cem_logs.csv', index=False)
```

```python
# plot training logs
plt.figure(figsize=[10,5])

plt.plot(logs['episode'], logs['elite_avg_timestamp_last_20_episodes'],'k', label='elite')
plt.plot(logs['episode'], logs['all_avg_timestamp_last_20_episodes'], 'm', label='general')
plt.legend()
plt.title('CEM Training: Elite mean vs Genetal mean timestamp, averaged over last 20 iterations')
plt.ylabel('timestamps')
plt.xlabel('iteration')
plt.ylim([0, logs['elite_avg_timestamp_last_20_episodes'].max() + 20])

plt.savefig('../figure/plot/cem_plot.jpg')
plt.show()
```
